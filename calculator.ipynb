```python
result=0
def main1():
        print("\n\t\t\t***Calculator app***\n")
        while True :
            # choices for user
            print("1.Addition")
            print("2.Subtraction")
            print("3.Multipliction")
            print("4.Division")
            print("5.Modulus")
            print("6.Fllor divisiond")
            print("7.Power")
            print("8.Exit")
            # taking input from the user
            choice =int(input("\nEnter your Choice : "))
            # for exiting 
            if(choice == 8):
                print("\n\n")
                break
            # for checking out of range choices
            elif(choice>8):
                print("\nInvalid choice Please, try again later\n")
            else:
                a = int(input("\nEnter the first Number : "))
                b = int(input("Enter the Second Number : "))

                if(choice == 1):
                    add(a,b)
                elif(choice == 2):
                    sub(a,b)
                elif(choice == 3):
                    mul(a,b)
                elif(choice == 4):
                    div(a,b)
                elif(choice == 5):
                    mod(a,b)
                elif(choice == 6):
                    flr(a,b)
                elif(choice == 7):
                    power(a,b)
                elif(choice == 8):
                    print("\nExiting from Calculator......\n")
                    break;
                else:
                    print("\nInvalid choice Please, try again later\n")

# add() method
def add(x,y):
    result=x+y
    print(f"\nAddition of {x} and {y} is : {result}")
    print("________________________________________")

 # sub() method   
def sub(x,y):
    result=x-y
    print(f"\nSubtraction of {x} and {y} is : {result}")
    print("________________________________________")

# mul() method
def mul(x,y):
    result=x*y
    print(f"\nMultiplication of {x} and {y} is : {result}")
    print("________________________________________")

# div() method
def div(x,y):
    if(y==0):
        print("\nCan't Divide by 'Zero'")
    else:
        result = x/y
        print(f"\nDivision of {x} and {y} is : {result}")
    print("________________________________________")

# mod() method
def mod(x,y):
    if(y == 0):
        print("\nCan't Divide by 'Zero'")
    else:
        result=x%y
        print(f"\nModulus of {x} and {y} is : {result}")
    print("________________________________________")
    
# flr() method
def flr(x,y):
    if(y == 0):
        print("\nCan't Divide by 'Zero'")
    else:
        result=x//y
        print(f"\nFloor Division of {x} and {y} is : {result}")
    print("________________________________________")

# power() method
def power(x,y):
    result=x**y
    print(f"\n{x} to the Power of {y} is : {result}")
    print("__________________________________________")

# main function
if __name__ == "__main__":
        main1()
        print("\n\t\tThanks for using..........\n")
```
